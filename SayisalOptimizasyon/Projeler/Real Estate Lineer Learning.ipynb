```python
np.random.seed(1)
weight=np.random.rand()
bias=np.random.rand()
```

```python
def fiyat_tahmin(m2):
    global weight,bias
    return m2*weight+bias
def hata_payi(m2,fiyat):
    tahmin=fiyat_tahmin(m2)
    return fiyat-tahmin
```

```python

plt.plot(data["living_area_renov"],fiyat_tahmin(data["living_area_renov"]),color="red")
sns.scatterplot(data= data, x="living_area_renov",y="Price",hue="Notater")
plt.show()
```

```python
epoch=100
learning_rate=1e-3
tahmin=fiyat_tahmin(data["living_area_renov"][0])
hata=hata_payi(data["living_area_renov"][0],data["Price"][0])
weight=weight+learning_rate*hata * data.loc[0]["living_area_renov"]
bias=bias+learning_rate*hata
weight,bias
```

```python


for j in range(epoch):
    for i in range(len(data)):
        tahmin=fiyat_tahmin(data["living_area_renov"][i])
        hata=hata_payi(data["living_area_renov"][i],data["Price"][i])
        weight=weight+learning_rate*hata * data.loc[i]["living_area_renov"]
        bias=bias+learning_rate*hata

    print(f"Epoch: {j}  , Weight: {weight}  , Bias: {bias}")
        
weight1=weight
bias1=bias
```
